```python


# This is  the colormap I'd like to use.
# cm = plt.cm.get_cmap('RdYlBu_r')
cm = plt.cm.get_cmap('PRGn')

#plot the data

fig, ax = plt.subplots(1,2,figsize=(10,7), gridspec_kw={'width_ratios': [5, 1]})
plt.subplot(121, projection="mollweide")
plt.grid(True)
plt.subplots_adjust(right=0.9, top=None, wspace=None, hspace=None)
plt.scatter(gal.galactic.l.wrap_at('180d').rad, gal.galactic.b.rad, marker='*', s =30 , c = zhel, cmap = cm)
# p lt.scatter(gal.galactic.l.wrap_at('180d').rad, gal.galactic.b.rad, marker='*', s =30 , c = zhel, cmap = cm, norm=matplotlib.colors.LogNorm())
cmap=plt.colorbar(label = r'$z_{hel}$', fraction=0.02, location = 'left', ticks = [0., 0.1, 1.0, 2.5], )
cmap.ax.set_yticklabels(['0','0.1', '1.0','2.5']) 
plt.tick_params(which='both', labelsize=14)
cmap.set_label(label=r'$z_{hel}$', size=20, weight='bold')
cmap.ax.tick_params(labelsize= 14)
plt.tick_params(which='both', labelsize=14)
plt.subplot(122, projection=None, aspect=2.5)

# Plot histogram.

n, bins, patches = plt.hist(zhel, 10, orientation = "horizontal")
print(bins)
# bins=(0.0001,0.25, 0.5, 0.75, 1., 2.25)
bin_centers = 0.5 * (bins[:-1] + bins[1:])
# plt.xscale('logit')# scale values to interval [0,1]
plt.xscale('log')
# plt.gca().yaxis.set_label_position("right")
col = bin_centers - min(bin_centers)
col /= max(col)
# cm(norm=matplotlib.colors.LogNorm()
for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))
plt.ylabel(r'$z_{hel}$', fontsize=20)
plt.xlabel(r'Number of SNe',fontsize=20)
plt.tick_params(which='both', labelsize=14)
plt.tight_layout()


plt.show()
```
